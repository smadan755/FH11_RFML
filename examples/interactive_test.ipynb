```python
import matlab.engine
import numpy as np
import matplotlib.pyplot as plt 

# Start MATLAB engine
eng = matlab.engine.start_matlab()
print("MATLAB engine started!")
```

```python
# run the bpsk function
eng.addpath(r'c:\Users\madan\FH11_RFML', nargout=0)
eng.addpath(r'c:\Users\madan\npy-matlab\npy-matlab', nargout=0)  # Add npy-matlab to path



data_samp_count = 1
save_path = r"C:\Users\madan\FH11_RFML\data_bpsk"



fs = 48e3
Tsymb = 1e-3
sps = fs*Tsymb
Nsymb = 2048
output_len = Nsymb*sps
fc = 6e3


#call bpsk_function - note the nargout=0 because bpsk() doesn't return anything
eng.bpsk(data_samp_count, save_path, output_len, fs, Tsymb, fc, nargout=0)
```

```python
data = np.load(r"C:\Users\madan\FH11_RFML\data_bpsk\bpsk_1.npy")

data = data.flatten()

T = len(data)/fs

t = np.linspace(0,T,len(data))


plt.figure(figsize=(12,6))
plt.plot(t,data)
plt.xlim([0,0.005])
```
